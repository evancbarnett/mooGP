```python
import time
import numpy as np
```

```python
import sys
import os

# Go up one level to the project root so Python can find 'moogp' and 'ogp'
sys.path.append(os.path.abspath('..'))

from moogp.model import MOOGP
```

```python

rng = np.random.default_rng(42)
# The Fast path dominates when n and p are large!
n, d, p, q = 300, 5, 20, 5  

X = rng.uniform(-1.0, 1.0, size=(n, d))
Y = rng.normal(size=(n, p))
data = {"X_scaled": X, "Y": Y}

sigma_eps2 = np.full(p, 0.1)

# We must give both models the EXACT same Psi to make it a fair math race
from scipy.linalg import svd
_, svals, Vt = svd(Y, full_matrices=False)
Phi = Vt.T[:, :q] * np.sqrt(n) / svals[:q]
Psi_fair = np.diag(np.sqrt(sigma_eps2)) @ Phi

theta = np.zeros(q * (d + 1)) # Dummy parameters

print(f"Matrix Size for Slow Path: {n*p} x {n*p}")

# --- SLOW PATH ---
model_slow = MOOGP(terms=[None], q=q, Psi=Psi_fair, learn_Psi=False, 
                   sigma_eps2=sigma_eps2, use_diagonalized_interaction=False)
model_slow._prepare_data(data)

start = time.perf_counter()
model_slow._nll(theta)
slow_time = time.perf_counter() - start
print(f"Slow Path 1-Pass Time: {slow_time:.4f} seconds")

# --- FAST PATH ---
model_fast = MOOGP(terms=[None], q=q, Psi=Psi_fair, learn_Psi=False, 
                   sigma_eps2=sigma_eps2, use_diagonalized_interaction=True)
model_fast._prepare_data(data)

start = time.perf_counter()
model_fast._nll(theta)
fast_time = time.perf_counter() - start
print(f"Fast Path 1-Pass Time: {fast_time:.4f} seconds")

print(f"\nSpeedup: {slow_time / fast_time:.1f}x faster per iteration!")

```
