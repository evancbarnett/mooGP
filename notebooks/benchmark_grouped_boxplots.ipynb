# Benchmark grouped box plots

Point this notebook at a single `results.csv`, choose which models to include, and it will render grouped box plots for train time, RMSE, and 95% coverage.

```python
from pathlib import Path
import sys

from IPython.display import display

candidates = [Path.cwd(), Path.cwd().parent]
REPO_ROOT = next(path for path in candidates if (path / "experiments").exists())
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

from experiments.plot_benchmark_boxplots import create_metric_figures
```

```python
RESULTS_CSV = REPO_ROOT / "results-time-test" / "results.csv"
MODELS = ["MOOGP", "MOGP", "LCGP", "OILMM"]
OUTPUT_DIR = None  # Example: REPO_ROOT / "results-time-test" / "plots"

RESULTS_CSV
```

```python
figures = create_metric_figures(
    results_csv=RESULTS_CSV,
    models=MODELS,
    output_dir=OUTPUT_DIR,
    show=False,
)

for metric in ("train_time_sec", "rmse", "coverage_95"):
    display(figures[metric])
```
